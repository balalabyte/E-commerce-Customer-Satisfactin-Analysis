## 2. Assume that, given the results of surveys, you want to predict the value of customer satisfaction (satis). Using your intuitions and insights, select 8-10 variables that you think can best predict customer satisfaction. Discuss the reason you choose those variables. (20%)

1. **Product Accuracy (product_accuracy)**: Accurate product descriptions and expectations that match the delivered product are crucial for customer satisfaction. When products meet or exceed expectations based on their online descriptions, customer satisfaction is likely to be higher.

2. **Information Sufficiency (info_sufficiency)**: It is important to provide all the information regarding the product for the customers to feel confident in purchasing the product and feel satisfied in with the process of online shopping.

3. **Delivery Timeliness (delivery_timeliness)**: On-time delivery is a critical component of customer satisfaction, especially in an e-commerce context where delivery expectations are set at the time of purchase.

4. **Website Security (website_security)**: With increasing concerns about online privacy and security, a secure website can significantly enhance customer trust and satisfaction.

5. **Return Policy (return_policy)**: A flexible and customer-friendly return policy can alleviate purchasing risks, leading to higher customer satisfaction by providing a safety net for online purchases.

6. **Checkout Ease (checkout_ease)**: A smooth and hassle-free checkout process minimizes frustration and abandonment at the final step of the purchasing process, directly impacting satisfaction.

7. **Navigation Ease (navigation_ease)**: The ease with which customers can find what they are looking for on a website directly affects their shopping experience and overall satisfaction.

8. **Promotions Attractiveness (promotions_attractiveness)**: Attractive promotions not only provide value but also enhance the shopping experience, leading to higher satisfaction levels.

9. **Feedback Availability (feedback_availability)**: The ability to easily provide feedback allows customers to feel heard and valued, which can improve their satisfaction with the company.

10. **Interest Variety (interest_variety)**: Offering a variety of products that cater to different interests can improve the likelihood that customers find what they are looking for, enhancing satisfaction.

These variables are chosen based on the premise that customer satisfaction is influenced by a mix of transactional experiences (such as checkout ease and delivery timeliness), post-purchase support (like return policies and feedback mechanisms), and the overall perception of the brand (captured by overall satisfaction and security concerns). Each of these factors contributes to the holistic experience a customer has with a company, thereby influencing their satisfaction.

````python
selected_features =['info_sufficiency','checkout_ease',
                    'website_security','delivery_range',
                     'order_tracking_ease','return_policy']
X= df[selected_features]
y= df['satis']
````

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
````

````python
#Mean imputation
# Impute missing values with mean for X and y
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X_imputed = imputer.fit_transform(X)
# reshaping to 2D, imputing, and then flattening back to 1D for scikit-learn modelling
y_imputed = imputer.fit_transform(y.values.reshape(-1, 1)).ravel()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y_imputed, test_size=0.2, random_state=42)

# Initialize and fit the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# SO we choose MSE as the error metric, Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

mse
````

Performing Recursive Feature Elimination (RFE) to select the best features for predicting customer satisfaction.

````python
features =['interest_variety','needs_variety','navigation_ease','find_ease','availability','promotions_attractiveness',
 'info_sufficiency','compare_ease','special_requests_ease','cart_management_ease','checkout_ease','website_security',
 'delivery_info_clarity','delivery_range','order_tracking_ease','delivery_timeliness','product_accuracy',
 'feedback_availability','return_policy']
````

````python
from sklearn.feature_selection import RFE

X= df[features]
y= df['satis']
model= LinearRegression()
X_train, X_test, y_train, y_test= train_test_split(X, y, random_state=0,train_size=0.8)

# Initialize RFE with the linear regression model and the desired number of features
# Let's say we want to select 6 features
rfe = RFE(estimator=model, n_features_to_select=6, step=1)

# Fit RFE
rfe.fit(X_train, y_train)

# now we do model prediction on test data
y_pred = rfe.predict(X_test)

# Calculate MSE to evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
````

````python
selected_features = X.columns[rfe.support_]
print("Selected features:", selected_features)
````

## 4. Use the linear model to determine which KPIs could most effectively improve customer satisfaction at eBay. (20%)


````python
final_model = LinearRegression()
final_model.fit(X_train[selected_features], y_train)

coefficients = pd.DataFrame(final_model.coef_, selected_features, columns=['Coefficient'])

coefficients.sort_values(by='Coefficient', ascending=False)
````

The KPIs are sorted in the order of having the most effect on customer satisfaction.
Return policy, Special requests , Feedback availability are the top 3.
